```python
count = -1
for tid in df["Tidspunkt_Faktisk_Avgang_Holdeplass_Fra"]:
    count+=1
    if tid.startswith('25'):
        print ("Time replaced :", tid.replace('25', '01'))
        print ("Counter to find index :", count)
        x = tid.split(":")
        nytid = x[0].replace('25', '01')
        nytid = x[0].replace('24', '00')
        nytid += (":") + x[1] + (":") + x[2]
        df["Tidspunkt_Faktisk_Avgang_Holdeplass_Fra"].loc[[count]] = nytid
        df['Tidspunkt_Faktisk_Avgang_Holdeplass_Fra'].loc[[count]] = df['Tidspunkt_Faktisk_Avgang_Holdeplass_Fra'].replace(df["Tidspunkt_Faktisk_Avgang_Holdeplass_Fra"].loc[[count]], nytid)
        df.at[count, 'Tidspunkt_Faktisk_Avgang_Holdeplass_Fra'] = nytid

        
        print ("Whats on this index :", df["Tidspunkt_Faktisk_Avgang_Holdeplass_Fra"].loc[[count]], "\n")
```
